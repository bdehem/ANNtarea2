```python
from keras.datasets import mnist
from keras.utils import np_utils
import numpy as np

def make_sets(nval=1000):
    (x_train, y_train), (x_test, y_test) = mnist.load_data()
    x_train = x_train.astype('float32') / 255.
    x_test = x_test.astype('float32') / 255.
    x_train = x_train.reshape((len(x_train), np.prod(x_train.shape[1:])))
    x_test = x_test.reshape((len(x_test), np.prod(x_test.shape[1:])))
    x_val = x_train[-nval:]
    y_val = y_train[-nval:]
    x_train = x_train[:-nval]
    y_train = y_train[:-nval]
    Y_train = np_utils.to_categorical(y_train, 10)
    Y_val = np_utils.to_categorical(y_val, 10)
    Y_test = np_utils.to_categorical(y_test, 10)
    return (x_train, y_train, Y_train), (x_test, y_test, Y_test), (x_val, y_val, Y_val)


(x_train, y_train, Y_train), (x_test, y_test, Y_test), (x_val, y_val, Y_val) = make_sets(1000)
```

```python
from keras.layers import Input, Dense
from keras.models import Model
from keras.optimizers import SGD

def reduce_dim(out_dim, encoder_activation = 'sigmoid'):
    input_img = Input(shape=(784,))
    encoded = Dense(out_dim, activation='sigmoid')(input_img)
    decoded = Dense(784, activation='sigmoid')(encoded)
    autoencoder = Model(input=input_img, output=decoded)
    encoder     = Model(input=input_img, output=encoded)
    encoded_input = Input(shape=(out_dim,))
    decoder_layer = autoencoder.layers[-1]
    decoder = Model(input=encoded_input, output=decoder_layer(encoded_input))
    autoencoder.compile(optimizer=SGD(lr=1.0), loss='binary_crossentropy')
    hist = autoencoder.fit(x_train,x_train,nb_epoch=10,batch_size=25,shuffle=True,validation_data=(x_val, x_val),verbose=0)
    #autoencoder.save('basic_autoencoder_768x'+ str(out_dim) +'.h5')
    #encoder.save('basic_encoder_768x'+ str(out_dim) +'.h5')
    #decoder.save('basic_decoder_768x'+ str(out_dim) +'.h5')
    err = hist.history['val_loss'][-1]
    return err
```

```python
def renderImages(encoder, decoder):
    encoded_test = encoder.predict(x_test)
    decoded_test = decoder.predict(encoded_test)
    n = 10
    plt.figure(figsize=(20, 4))

    for i in range(n):
        ax = plt.subplot(2, n, i + 1)
        plt.imshow(x_test[i].reshape(28, 28))
        plt.gray()
        ax.get_xaxis().set_visible(False)
        ax.get_yaxis().set_visible(False)
        ax = plt.subplot(2, n, i + 1 + n)
        plt.imshow(decoded_test[i].reshape(28, 28))
        plt.gray()
        ax.get_xaxis().set_visible(False)
        ax.get_yaxis().set_visible(False)
    plt.show()
    return
```

```python
from sklearn.neighbors import KNeighborsClassifier

def KNeighborsAccuracy (encoder):
    encoded_train = encoder.predict(x_train)
    encoded_test = encoder.predict(x_test)

    clf = KNeighborsClassifier(10)
    clf.fit(encoded_train, Y_train)
    clf.fit(encoded_train, Y_train)
    score = clf.score(encoded_test,Y_test)

    return score
```

```python
def clusteringAccuracy(pred_labels, y, nclusters=10):
    true_pred = 0.0
    for i in range(0,nclusters):
        mvlabel = np.argmax(np.bincount(y[pred_labels==i]))
        true_pred += sum(y[pred_labels==i] == mvlabel)
        return true_pred/len(y)

from sklearn.cluster import KMeans
from sklearn import metrics

def clusteringAccuracyMeasurement (encoder):
    encoded_train = encoder.predict(x_train)
    encoded_test = encoder.predict(x_test)

    model = KMeans(n_clusters=10)
    labels_pred = model.fit_predict(encoded_train)
    score = metrics.adjusted_rand_score(y_train, labels_pred)
    
    print '  ARI: %.2f' % score
    print '  CC:  %.2f' % clusteringAccuracy(labels_pred, y_train)
    
    return
```

```python
print ("Clustering Accuracy con 2 neuronas:" )
clusteringAccuracyMeasurement (encoder2)
print ("Clustering Accuracy con 8 neuronas:" )
clusteringAccuracyMeasurement (encoder8)
print ("Clustering Accuracy con 32 neuronas:")
clusteringAccuracyMeasurement (encoder32)
print ("Clustering Accuracy con 64 neuronas:")
clusteringAccuracyMeasurement (encoder64)
```

```python
from sklearn.decomposition import PCA
from sklearn.neighbors import KNeighborsClassifier

def pcaTrain(pca):
    pca.fit(x_train)
    pca_train = pca.transform(x_train)
    pca_test = pca.transform(x_test)
    clf = KNeighborsClassifier(10)
    clf.fit(pca_train, y_train)
    clfScore = clf.score(pca_test,y_test)
    return clf.score(pca_test,y_test)

def pcaClassifyAccuracy (pca):
    encoded_train = pca.transform(x_train)
    encoded_test  = pca.transform(x_test)

    clf = KNeighborsClassifier(10)
    clf.fit(encoded_train, Y_train)
    clf.fit(encoded_train, Y_train)
    score = clf.score(encoded_test,Y_test)

    return score

def pcaClusteringAccuracyMeasurement (pca):
    encoded_train = pca.transform(x_train)
    encoded_test  = pca.transform(x_test)

    model = KMeans(n_clusters=10)
    labels_pred = model.fit_predict(encoded_train)
    score = metrics.adjusted_rand_score(y_train, labels_pred)
    
    print '    ARI: %.2f' % score
    print '    CC:  %.2f' % clusteringAccuracy(labels_pred, y_train)
    
    return
```

```python
pca2  = PCA (n_components=2 )
pca8  = PCA (n_components=8 )
pca32 = PCA (n_components=32)
pca64 = PCA (n_components=64)


print 'PCA with 2 components:'
print '  clf score %.2f' % pcaTrain(pca2 )
print '  Knn score %.2f' % pcaClassifyAccuracy(pca2 )
print '  Kmn score:'
pcaClusteringAccuracyMeasurement(pca2 )
print 'PCA with 8 components:'
print '  clf score %.2f' % pcaTrain(pca8 )
print '  Knn score %.2f' % pcaClassifyAccuracy(pca8 )
print '  Kmn score:'
pcaClusteringAccuracyMeasurement(pca8 )
print 'PCA with 32 components:'
print '  clf score %.2f' % pcaTrain(pca32)
print '  Knn score %.2f' % pcaClassifyAccuracy(pca32)
print '  Kmn score:'
pcaClusteringAccuracyMeasurement(pca32)
print 'PCA with 64 components:'
print '  clf score %.2f' % pcaTrain(pca64)
print '  Knn score %.2f' % pcaClassifyAccuracy(pca64)
print '  Kmn score:'
pcaClusteringAccuracyMeasurement(pca64)
```

```python
from sklearn.neural_network import BernoulliRBM

import numpy as np
import pickle ##to save trained models

def rbmTrain (rbm, n):
    rbm.fit(x_train) ##Train using persistent Gibbs chains
    
    fileo = open('basicRBM' + str(n) + '.pickle','wb')
    pickle.dump(rbm, fileo)
    fileo.close()
    
    print('RBM n: ' + str(n) + ' trained and saved!')
    
    return

def rbmTest (rbm):
    return rbm.score_samples(x_test).mean() # pseudo-likelyhood mean

def rbmClassifyAccuracy (rbm):
    encoded_train = rbm.transform(x_train)
    encoded_test  = rbm.transform(x_test)

    clf = KNeighborsClassifier(10)
    clf.fit(encoded_train, Y_train)
    clf.fit(encoded_train, Y_train)
    score = clf.score(encoded_test,Y_test)

    return score

def rbmClusteringAccuracyMeasurement (rbm):
    encoded_train = rbm.transform(x_train)
    encoded_test  = rbm.transform(x_test)

    model = KMeans(n_clusters=10)
    labels_pred = model.fit_predict(encoded_train)
    score = metrics.adjusted_rand_score(y_train, labels_pred)
    
    print '    ARI: %.2f' % score
    print '    CC:  %.2f' % clusteringAccuracy(labels_pred, y_train)
    
    return
```

```python
rbm2  = BernoulliRBM(n_components=2,  batch_size=25, learning_rate=0.05, verbose=0, n_iter=50)
rbm8  = BernoulliRBM(n_components=8,  batch_size=25, learning_rate=0.05, verbose=0, n_iter=50)
rbm32 = BernoulliRBM(n_components=32, batch_size=25, learning_rate=0.05, verbose=0, n_iter=50)
rbm64 = BernoulliRBM(n_components=64, batch_size=25, learning_rate=0.05, verbose=0, n_iter=50)

rbmTrain(rbm2,  2 )
rbmTrain(rbm8,  8 )
rbmTrain(rbm32, 32)
rbmTrain(rbm64, 64)
```

```python
print 'RBM with 2 components:'
print '  clf score %.2f' % rbmTest(rbm2 )
print '  Knn score %.2f' % rbmClassifyAccuracy(rbm2 )
print '  Kmn score:'
rbmClusteringAccuracyMeasurement(rbm2 )
print 'RBM with 8 components:'
print '  clf score %.2f' % rbmTest(rbm8 )
print '  Knn score %.2f' % rbmClassifyAccuracy(rbm8 )
print '  Kmn score:'
rbmClusteringAccuracyMeasurement(rbm8 )
print 'RBM with 32 components:'
print '  clf score %.2f' % rbmTest(rbm32)
print '  Knn score %.2f' % rbmClassifyAccuracy(rbm32)
print '  Kmn score:'
rbmClusteringAccuracyMeasurement(rbm32)
print 'RBM with 64 components:'
print '  clf score %.2f' % rbmTest(rbm64)
print '  Knn score %.2f' % rbmClassifyAccuracy(rbm64)
print '  Kmn score:'
rbmClusteringAccuracyMeasurement(rbm64)
```
